```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.exp(-np.power(x, 2))
```

```python
sample_size = 10_000_000
sample = np.random.random((sample_size, 2))
min_x, max_x = -2, 2
min_y, max_y = f(-2), f(0)
box = sample * np.abs([max_x - min_x, max_y]) + [min_x, 0]
box_size = np.abs(max_x - min_x) * np.abs(max_y)
xs, ys = np.split(box, [1], axis=1)
under_curve = np.sum(ys <= f(xs))
size_inder_curve = under_curve / sample_size * box_size
print(f'total area: {box_size}')
print(f'points under the curve: {under_curve}')
print(f'computed integral: {size_inder_curve}')
```

```python
plt.figure(figsize=(15, 10))
lin = np.linspace(min_x, max_x, 1_000_000)
lin_y = f(lin)
plt.scatter(xs, ys)
plt.plot(lin, lin_y, 'r')
plt.show()
```
